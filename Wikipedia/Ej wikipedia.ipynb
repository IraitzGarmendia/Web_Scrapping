```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://en.wikipedia.org/wiki/Forbes_list_of_the_world%27s_highest-paid_athletes#2023_list"
```

```python
text_web = requests.get(URL).text
print(text_web)
```

#### Le damos formato de lxml

```python
sopa = BeautifulSoup(text_web, "lxml")
print(sopa)
```

```python
print(sopa.prettify())
```

```python
#Para ver el titulo entero
print(sopa.title)
#Aqui nos devuelve solo el titulo
print(sopa.title.string)
```

```python
#Para ver los parrafos que tengo y buscas todo los parrafos (p) como una lista
sopa.find_all("p")
```

```python
sopa.find_all("h2")
```

```python
sopa.find_all("table")  #Los documentos que tengo como tabla
```

```python
#Encuentrame todas las tablas pero dentro de la clase de wikitable sortable
tablas = sopa.find_all("table", class_ = "wikitable sortable")
print(tablas)
```

```python
import pandas as pd
```

```python
wiki = pd.read_html(str(tablas))
```
